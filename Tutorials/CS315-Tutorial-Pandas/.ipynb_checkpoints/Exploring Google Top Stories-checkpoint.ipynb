```python
import os
import json
import pandas as pd
from urllib.parse import urlparse
```

```python
def extract_domain(url): #domain from the url
    parsed_url = urlparse(url)
    return parsed_url.netloc
    
def process_folder(folder_name):
    data = []
    for root, dirs, files in os.walk(folder_name):
        for file in files:
            if file.endswith(".json") and not file.startswith("."):
                file_path = os.path.join(root, file)
                with open(file_path, "r") as f:
                    json_data = json.load(f)
                    for item in json_data:
                        domain = extract_domain(item["url"])
                        data.append({"title": item["title"], "domain": domain})   #asked Chat GPT about the syntax, could not figure out how to add data to dataframe
    return data
```

```python
data = process_folder("GoogleTopStories")
df = pd.DataFrame(data)
csv = df.to_csv("GoogleTopStories.csv", index=False)
```
